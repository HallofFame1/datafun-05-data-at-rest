# SECTION 1. Import CSV Data

```python
import pandas as pd

pd.options.display.max_rows = 10

clients_data = pd.read_csv('clients.csv', index_col=['num'])

# print(clients_data.to_string())

n = 3
# Select first n columns
first_n_columns  = clients_data.iloc[: , :n]

print(first_n_columns)
```

# SECTION 2. Clean Data

### Renaming some columns

```python
clients_data.rename(columns = {'chastity_number':'client_number', 'sub_county':'district', 'mfl_code':'insurance_balance'}, inplace = True)
clients_data.head()
```

### Replacing NaN data values

```python
import numpy as np
clients_data = clients_data.replace(np.nan, 'Missing')
clients_data.head()
```

### Dropping some columns

```python
columns = ['date_of_enrollment', 'modality', 'current_county_of_residence_othe', 'rslrf_viral_load_lab_name']

clients_data = pd.DataFrame(clients_data).drop(columns, axis=1)

clients_data.head()
```
